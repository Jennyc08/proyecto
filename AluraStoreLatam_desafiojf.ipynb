#1. Análisis de facturación



```python
# Importar pandas y cargar las tiendas
import pandas as pd

# URLs de las bases de datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Verificar nombres de columnas
print("Columnas tienda1:", tienda1.columns.tolist())
print("Columnas tienda2:", tienda2.columns.tolist())
print("Columnas tienda3:", tienda3.columns.tolist())
print("Columnas tienda4:", tienda4.columns.tolist())
```

```python
# Calcular la facturación total de cada tienda usando la columna 'Precio'
facturacion_1 = tienda1['Precio'].sum()
facturacion_2 = tienda2['Precio'].sum()
facturacion_3 = tienda3['Precio'].sum()
facturacion_4 = tienda4['Precio'].sum()

# Mostrar resultados con formato
print(f"Facturación Tienda 1: ${facturacion_1:,.2f}")
print(f"Facturación Tienda 2: ${facturacion_2:,.2f}")
print(f"Facturación Tienda 3: ${facturacion_3:,.2f}")
print(f"Facturación Tienda 4: ${facturacion_4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# -------- Gráfico 1: Facturación total por tienda (barras) --------
facturaciones = [
    tienda1['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(8, 5))
plt.bar(tiendas, facturaciones)
plt.title("Facturación Total por Tienda")
plt.ylabel("Ingresos ($)")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
# Cantidad de productos vendidos por categoría en cada tienda

print("Tienda 1 - Ventas por categoría:")
print(tienda1.groupby('Categoría del Producto').size().sort_values(ascending=False))

print("\nTienda 2 - Ventas por categoría:")
print(tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False))

print("\nTienda 3 - Ventas por categoría:")
print(tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False))

print("\nTienda 4 - Ventas por categoría:")
print(tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False))
```

```python
# -------- Gráfico 2: Distribución de productos por categoría (torta) --------
# Uniremos todas las tiendas para un panorama general
todas = pd.concat([tienda1, tienda2, tienda3, tienda4])
categorias = todas['Categoría del Producto'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribución de Productos Vendidos por Categoría")
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular calificación promedio por tienda

promedio_calif_1 = tienda1['Calificación'].mean()
promedio_calif_2 = tienda2['Calificación'].mean()
promedio_calif_3 = tienda3['Calificación'].mean()
promedio_calif_4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"Calificación promedio - Tienda 1: {promedio_calif_1:.2f}")
print(f"Calificación promedio - Tienda 2: {promedio_calif_2:.2f}")
print(f"Calificación promedio - Tienda 3: {promedio_calif_3:.2f}")
print(f"Calificación promedio - Tienda 4: {promedio_calif_4:.2f}")
```

```python
# -------- Gráfico 3: Calificación promedio por tienda (línea) --------
calificaciones = [
    tienda1['Calificación'].mean(),
    tienda2['Calificación'].mean(),
    tienda3['Calificación'].mean(),
    tienda4['Calificación'].mean()
]

plt.figure(figsize=(8, 5))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', linewidth=2)
plt.title("Calificación Promedio por Tienda")
plt.ylabel("Calificación (0 a 5)")
plt.xlabel("Tienda")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def mostrar_ventas_productos(tienda, nombre_tienda):
    conteo = tienda['Producto'].value_counts()

    # Mostrar en consola
    print(f"\n{nombre_tienda} - Top 5 productos más vendidos:")
    print(conteo.head(5))

    print(f"\n{nombre_tienda} - Top 5 productos menos vendidos:")
    print(conteo.tail(5))

    # Visualización
    plt.figure(figsize=(10, 4))
    conteo.head(5).plot(kind='bar')
    plt.title(f'{nombre_tienda} - Productos Más Vendidos')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(10, 4))
    conteo.tail(5).plot(kind='bar', color='gray')
    plt.title(f'{nombre_tienda} - Productos Menos Vendidos')
    plt.xlabel('Producto')
    plt.ylabel('Cantidad Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Ejecutar para cada tienda
mostrar_ventas_productos(tienda1, "Tienda 1")
mostrar_ventas_productos(tienda2, "Tienda 2")
mostrar_ventas_productos(tienda3, "Tienda 3")
mostrar_ventas_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt

# Calcular el costo de envío promedio por tienda
envio_promedio_1 = tienda1['Costo de envío'].mean()
envio_promedio_2 = tienda2['Costo de envío'].mean()
envio_promedio_3 = tienda3['Costo de envío'].mean()
envio_promedio_4 = tienda4['Costo de envío'].mean()

# Crear listas para graficar
envios = [envio_promedio_1, envio_promedio_2, envio_promedio_3, envio_promedio_4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, envios)
plt.title("Costo de Envío Promedio por Tienda")
plt.ylabel("Costo Promedio de Envío ($)")
plt.xlabel("Tienda")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Añadir columna para identificar de qué tienda proviene cada fila
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los datos
datos = pd.concat([tienda1, tienda2, tienda3, tienda4])
```

```python
# Crear gráfico de dispersión
plt.figure(figsize=(10, 6))

for tienda in datos['Tienda'].unique():
    df = datos[datos['Tienda'] == tienda]
    plt.scatter(df['lon'], df['lat'], label=tienda, alpha=0.5, s=10)

plt.title("Distribución Geográfica de Ventas por Tienda")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
